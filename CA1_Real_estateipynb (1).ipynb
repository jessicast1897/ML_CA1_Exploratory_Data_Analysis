```python
#data preprocessing
```

```python
#copy of data
df_clean=df.copy()
```

```python
#check for duplicates
dup_count = df_clean.duplicated().sum()
print("Duplicate rows:", dup_count)
if dup_count > 0:
    df_clean = df_clean.drop_duplicates()
```

```python
#serparate feature types
numeric_cols = df_clean.select_dtypes(include=["int64", "float64"]).columns
categorical_cols = df_clean.select_dtypes(include=["object"]).columns
```

```python
#missing value report
missing = df_clean.isna().sum()
missing_pct = (missing / len(df_clean)) * 100
missing_report = (pd.DataFrame({"missing": missing, "missing_pct": missing_pct}).sort_values("missing_pct", ascending=False))
missing_report
```

```python
#drop columns with lots of missing values
miss_thresh = 60
cols_drop = missing_report[missing_report["missing_pct"]>miss_thresh].index.tolist()
#dont drop sale price
cols_to_drop = [c for c in cols_drop if c != target]
cols_to_drop
```

```python
#must drop the above
```

```python
df_clean=df_clean.drop(columns=cols_to_drop)
```

```python
#recalc columns lists
numeric_cols = df_clean.select_dtypes(include=["int64", "float64"]).columns
categorical_cols = df_clean.select_dtypes(include=["object"]).columns
```

```python
#fix missing values numerical
num_fill_cols = [c for c in numeric_cols if c != target]
for c in num_fill_cols:
    if df_clean[c].isna().sum() > 0:
        df_clean[c] = df_clean[c].fillna(df_clean[c].median())
```

```python
#fix categorical
for c in categorical_cols:
    if df_clean[c].isna().sum() > 0:
        df_clean[c] = df_clean[c].fillna("Unknown")
```

```python
print("Missing values after imputation:", df_clean.isna().sum().sum())
```

```python
#checking for other ID columns
id_like = []
for c in df_clean.columns:
    if c != target:
        if df_clean[c].nunique() >= 0.99 * len(df_clean):
            id_like.append(c)

print("ID-like columns:", id_like)
```

```python
#one hot encoding
y = df_clean[target]
X = df_clean.drop(columns=[target])

X_encoded = pd.get_dummies(X, columns=X.select_dtypes(include=["object"]).columns, drop_first=True)

df_ready = pd.concat([X_encoded, y], axis=1)
```

```python
df_ready.head()
```

```python
#ID column is now an index
```

```python
#3 collinearity and feature selection (no class imbalance)
```

```python
df_part3=df_ready.copy()
```

```python
#correlation matrix
numeric_cols = df_part3.select_dtypes(include=["int64", "float64"]).columns

corr = df_part3[numeric_cols].corr()

plt.figure(figsize=(14, 10))
sns.heatmap(corr, center=0)
plt.title("Correlation Heatmap (All Numeric Features)")
plt.show()
```

```python
#identify correlation threshold
CORR_THRESHOLD = 0.90
feature_cols = [c for c in numeric_cols if c != target]
corr_features = df_part3[feature_cols].corr().abs()
```

```python
upper = corr_features.where(np.triu(np.ones(corr_features.shape), k=1).astype(bool))

high_corr_pairs = (
    upper.stack()
    .reset_index()
    .rename(columns={"level_0": "Feature_A", "level_1": "Feature_B", 0: "AbsCorr"})
    .sort_values("AbsCorr", ascending=False)
)

high_corr_pairs = high_corr_pairs[high_corr_pairs["AbsCorr"] >= CORR_THRESHOLD]

print(f"Highly correlated feature pairs (|corr| >= {CORR_THRESHOLD}):")
display(high_corr_pairs.head(30))
```
